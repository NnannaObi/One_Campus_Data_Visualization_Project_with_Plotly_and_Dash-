```python
import pandas as pd
from scipy.stats import zscore

df1= df.groupby(['Store'])['Weekly_Sales'].agg(['sum'])
df1 = df1.reset_index()
df1 = df1.rename(columns={'sum': 'Total_sales'}) # Rename the column
df1.head()
# Calculate the z-score for the 'Total_weekly_sales' column
df1['Z_Score'] = zscore(df1['Total_sales'])

# Output the DataFrame with the Z_Score column
df1.head()
```
